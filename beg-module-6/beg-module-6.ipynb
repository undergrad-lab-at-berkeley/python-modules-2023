To plot a function, we import the pyplot package from matplotlib (already done for you). From the pyplot package, we use the 'plt.plot' command. 

Look at the cells below to understand how it works.

Assume I have a function  $y=e^{-x/10} sin(x)$. I wish to plot the function $y=e^{-x/10} sin(x)$ for x values in the range [0, 20]. 

<b> Step 1: </b> Create an array of x values.

```python
x = np.arange(0, 20)
```

<b> Step 2: </b>
Write the function in terms of x.

```python
y = np.exp(-x/10)*np.sin(x)
```

<b> Step 3:</b> Use the plt.plot command. The x values go as the first argument and the y values go as the second argument.

```python
plt.plot(x,y)
```

Can you guess why the graph above looks so ugly?

It is because the values of x we fed to the function were generated by us in an interval of [0,20] using the np.arange function. np.arange function has a default step size of 1. The 'plt.plot' function above only has 20 data points (x values) to plot.

For a smooth graph, we need more data points. I could do this by modifying the 'np.arange' function and decreasing the step size. But there's an alternate an easier way to do the same. <b> Use 'np.linspace'. </b> Run the cell below to revise what the 'np.linspace' function does

```python
np.linspace?
```

Now, if I want 10000 evenly-spread data points of x in the interval [0, 20], I can do the following:

```python
new_x = np.linspace(0, 20, 10000)
```

Run the cells below to see an improved version of our graph.

```python
y = np.exp(-new_x/10)*np.sin(new_x)
plt.plot(new_x, y)
```
